```python
def get_sum(arr):
  sum = 0
  for j in range(len(arr)):
    if arr[j] == '.':
      continue

    sum += int(arr[j])
    if j == len(arr)-1:
      print(f"{sum}", end = " ")
  return sum
```

```python
def scientific_to_float(exponential):
  split_word = 'e'
  e_index = exponential.index('e')
  base = float(exponential[:e_index])
  exponent = float(exponential[e_index + 1:])
  float_number = base * (10 ** exponent)
  return "%.*f" % (100000, float_number)
```

```python
def generate_double(i):
  c = 0 
  while c<24:
    arr = str(i)
    print(f"{i} = ", end = "")
    sum = get_sum(arr)
    while(sum > 9):
      print(" = ", end=" ")
      arr = str(sum)
      sum = get_sum(arr)
    print()
    i *= 2
    c += 1
    if c%6 == 0:
      print("#############################################################")
```

```python
def generate_half(i):
  c = 0 
  while c<24:
    try:
      arr = scientific_to_float(str(i))
      # arr = remove_exponent(str(i))
    except:
      arr = str(i)
    print(f"{i} = ", end = "")
    arr = str(arr)
    sum = get_sum(arr)
    while(sum > 9):
      print(" = ", end=" ")
      arr = str(sum)
      sum = get_sum(arr)
    print()
    i /= 2
    c += 1
    if c%6 == 0:
      print("#############################################################")
```

```python
generate_double(1)
```

```python
generate_half(1)
```
